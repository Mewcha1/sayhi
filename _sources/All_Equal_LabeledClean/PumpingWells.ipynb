### Pumping wells 

##### Collaboration and co-design process with key stakeholders

The modeling team developed a simulation optimization procedure from a purely hydrologic perspective regardless of the social context. Thus, assumptions about decision variables, initial conditions, observation points, among other settings were set based on research team general modeling experience, rather than hands-on experience about the aquifer management. For example, our decision variables are either selected pumping wells or pumping clusters.  To optimize withdrawal of the 92 pumping wells in the study area results in a high dimensional problem. This is computationally unfeasible since the iteration size  and the number of iterations to reach convergence increase with increasing the number of decision variables. Also, having a high dimensional problem is not technically recommended since large number of decision variables can cause solution non-uniqueness resulting in multimodality. Therefore, we grouped the 92 pumping wells into 20 clusters based on spatial proximity using k-mean clustering method.
